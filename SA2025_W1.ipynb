### 1. What is name of car that has the highest horsepower?  

```python
cars_data = pd.read_csv('/content/Cars.csv')
car_with_max_hp = cars_data.loc[cars_data['horsepower'].idxmax(), 'name']
print("Car with the highest horsepower:", car_with_max_hp)
```

### 2. How many cars have mpg ≥ 35?

```python
high_mpg_cars = cars_data[cars_data['mpg'] >= 35]
count = len(high_mpg_cars)
print(f"Number of cars with mpg ≥ 35: {count}")
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
filtered_cars = cars_data[(cars_data['horsepower'] > 100) & (cars_data['weight'] < 3000)]
most_common_origin = filtered_cars['origin'].mode()[0]
print("Most common origin for cars with horsepower > 100 and weight < 3000:", most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
for origin in cars_data['origin'].unique():
    origin_cars = cars_data[cars_data['origin'] == origin]
    mean_accel = round(origin_cars['acceleration'].mean(), 2)
    print(f"Mean acceleration for {origin}: {mean_accel}")
```

### 5. Which year had the highest average mpg?

```python
avg_mpg_by_year = cars_data.groupby('model_year')['mpg'].mean()
best_year = avg_mpg_by_year.idxmax()
highest_avg_mpg = round(avg_mpg_by_year.max(), 2)
print(f"Year with highest average MPG: {best_year} (Avg MPG: {highest_avg_mpg})")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
median_mpg = cars_data['mpg'].median()
cars_data['hp_to_weight'] = cars_data['horsepower'] / cars_data['weight']
high_mpg_cars = cars_data[cars_data['mpg'] > median_mpg]
best_ratio = high_mpg_cars['hp_to_weight'].max()
best_cars = high_mpg_cars[high_mpg_cars['hp_to_weight'] == best_ratio]
print(f"Best horsepower-to-weight ratio among high-MPG cars: {best_ratio:.4f}")
print("\nCars with this ratio:")
print(best_cars[['name', 'mpg', 'horsepower', 'weight', 'hp_to_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import matplotlib.pyplot as plt
import seaborn as sns

avg_mpg = cars_data.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=avg_mpg,
    x='model_year',
    y='mpg',
    hue='origin',
    marker='o',
    linewidth=2.5
)

plt.title('Evolution of Average MPG by Origin (1970-1982)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

X = horsepower

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
import pandas as pd

car_counts = cars_data['name'].value_counts()
multi_year_cars = car_counts[car_counts > 1].index.tolist()
multi_year_data = cars_data[cars_data['name'].isin(multi_year_cars)]
consistency_stats = (multi_year_data.groupby('name')['mpg']
                     .agg(['count', 'std', 'mean'])
                     .reset_index()
                     .rename(columns={'count': 'appearances',
                                     'std': 'mpg_std',
                                     'mean': 'avg_mpg'}))
consistent_cars = consistency_stats[
    (consistency_stats['mpg_std'] < 1.0) &
    (consistency_stats['appearances'] >= 2)
]
consistent_cars_sorted = consistent_cars.sort_values(
    by=['appearances', 'avg_mpg'],
    ascending=[False, False]
)
print("Most consistent car models (lowest MPG variation):")
print(consistent_cars_sorted[['name', 'appearances', 'avg_mpg', 'mpg_std']].to_string(index=False))
```
